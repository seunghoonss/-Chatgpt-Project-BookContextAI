```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, StaleElementReferenceException
import pandas as pd
import os
import requests
import re

# ChromeOptions 설정
op = webdriver.ChromeOptions()
op.add_argument('--headless')
op.add_argument('--no-sandbox')
op.add_argument('--disable-dev-shm-usage')

# WebDriver 초기화
driver = webdriver.Chrome(options=op)
wait = WebDriverWait(driver, 0.1)  # 로딩 시간을 0.1초로 변경

def save_image_from_url(url, save_path):
    """이미지 URL에서 이미지를 다운로드하고 지정된 경로에 저장하는 함수"""
    # 저장할 디렉토리가 존재하는지 확인하고, 없다면 생성
    directory = os.path.dirname(save_path)
    if not os.path.exists(directory):
        os.makedirs(directory)

    response = requests.get(url)
    with open(save_path, 'wb') as f:
        f.write(response.content)

def clean_filename(filename):
    s = re.sub(r'[\\/*?:"<>|]', '', filename)  # 허용되지 않는 문자 제거
    return s.replace(' ', '_')  # 공백을 언더스코어대신 대체

def get_element_data(css_selector, element_type="text"):
    try:
        if element_type == "text":
            return wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, css_selector))).text
        elif element_type == "attribute":
            return wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, css_selector))).get_attribute("value")
    except (TimeoutException, NoSuchElementException, StaleElementReferenceException):
        return "N/A"

# 첨부된 파일에서 아이템 ID와 추가 정보를 읽어옴
df_source = pd.read_excel("/content/drive/MyDrive/ChatGpt 프로젝트/코딩/kogpt2/Crawling Data/DB/정치 사회/교보문고_카테고리_상품리스트.xlsx") # path 경로 설정, 각 카테고리 별 Excel 파일 로드
df_source = df_source[:]

# YYYYMMDD 형식의 날짜를 YYYY년 MM월 DD일 형식으로 변환
def reformat_date(date):
    date = str(date)  # 날짜를 문자열로 변환
    return f"{date[:4]}년 {int(date[4:6])}월 {int(date[6:])}일"

# 크롤링 함수 정의
def crawl_kyobobook(item_id, additional_info):
    url = f"https://product.kyobobook.co.kr/detail/{item_id}"
    driver.get(url)

    # 각 데이터의 CSS 선택자 혹은 XPath를 사용하여 데이터 추출
    title = get_element_data("span.prod_title")
    author = get_element_data(".author a:first-child")
    editor_choice = get_element_data("div.info_text.fw_bold")
    book_intro = get_element_data("div.info_text:not(.fw_bold)")
    contents = get_element_data("li.book_contents_item")
    best_link_text = get_element_data("a.btn_best_link").replace("\n", ": ")
    price = get_element_data("span.prod_info_price > span.val") + "원"
    rating = get_element_data("input.form_rating.rating-input", "attribute") + "점"
    good_response = get_element_data("p.text_good_respon").replace('<br>', ' ')
    purchase_link = url
    genre1 = ' > '.join([elem.text for elem in driver.find_elements(By.CSS_SELECTOR, "li.category_list_item:first-child a")])
    genre2 = ' > '.join([elem.text for elem in driver.find_elements(By.CSS_SELECTOR, "li.category_list_item:last-child a")])
    image_url = driver.find_element(By.CSS_SELECTOR, 'img[src^="https://contents.kyobobook.co.kr/sih/fit-in"]').get_attribute("src")

    # 댓글들을 모두 가져옴
    comment_elements = driver.find_elements(By.CSS_SELECTOR, "div.comment_text")
    valid_comments = [elem.text for elem in comment_elements if elem.text.strip() != ""]

    # 상위 3개의 댓글만 선택하고, 각 댓글 앞에 "구매리뷰 {번호}:"를 추가
    first_three_comments = [f"구매리뷰 {i+1}: {text.split('.')[0]}." for i, text in enumerate(valid_comments[:3]) if text]
    combined_comments = "\n".join(first_three_comments)


    # 이미지 다운로드 및 저장
    clean_title = clean_filename(title)
    image_filename = f"{clean_title}.jpg"

    # 이미지 저장할 시, ''' ''' 주석 제거
    '''
    image_save_path = f"/content/drive/MyDrive/ChatGpt 프로젝트/코딩/kogpt2/Crawling Data/이미지(책표지)/정치 사회/{image_filename}" # path 설정
    save_image_from_url(image_url, image_save_path)
    '''
    cleaned_image_filename = image_filename.split('/')[-1].replace('.jpg', '').replace('_', ' ')

    # 크롤링한 정보와 추가 정보를 결합
    book_data = {
        "제목": title,
        "저자": author,
        "출판사": additional_info["출판사"],
        "편집장의 선택": editor_choice,
        "책소개": book_intro,
        "목차": contents,
        "주간베스트": best_link_text,
        "평점": rating,
        "구매자 응답": good_response,
        "구매리뷰": combined_comments,
        "구매링크": purchase_link,
        "장르1": genre1,
        "장르2": genre2,
        "정가": additional_info["정가"],
        "판매가": price,
        "할인율": additional_info["할인율"],
        "적립율": additional_info["적립율"],
        "적립예정포인트": additional_info["적립예정포인트"],
        "출판사": additional_info["출판사"],
        "발행(출시)일자": reformat_date(additional_info["발행(출시)일자"]),
        "이미지 파일명": cleaned_image_filename,
    }

    return book_data

# 각 아이템 ID에 대해 크롤링 수행
results = []
total_items = len(df_source)
for index, row in enumerate(df_source.iterrows()):
    current_count = index + 1
    remaining_count = total_items - current_count

    print(f"현재 크롤링 중: {current_count} / 남은 횟수: {remaining_count} / 총 횟수: {total_items}")

    item_id = row[1]["판매상품ID"]
    additional_info = {
        "정가": row[1]["정가"] + "원",
        "할인율": row[1]["할인율"] + "할인",
        "적립율": row[1]["적립율"] + "적립",
        "적립예정포인트": row[1]["적립예정포인트"] + "원",
        "출판사": row[1]["출판사"],
        "발행(출시)일자": row[1]["발행(출시)일자"]
    }
    result = crawl_kyobobook(item_id, additional_info)
    results.append(result)

# 데이터프레임 생성
df = pd.DataFrame(results)
display(df)
print()

# 저장할 파일명, 경로 저장
filename = "2_1th_kyobobook_data_combined.csv" #  path 설정
path = f"/content/drive/MyDrive/ChatGpt 프로젝트/코딩/kogpt2/Crawling Data/{filename}"
df.to_csv(path, index=False)

print(f"'{filename}' 파일이 성공적으로 저장되었습니다.")
```

```python
# 공백 포함 -> 공백 제거 -> 전체 글자로 판단

import pandas as pd
from IPython.display import display, HTML

# CSV 파일 로드
df = pd.read_csv('/content/drive/MyDrive/ChatGpt 프로젝트/코딩/kogpt2/Crawling Data/kyobobook_data5.csv')

# 이미지를 표시하는 HTML 문자열 생성
html_str = ""

for index, row in df.iterrows():
    # 이미지 파일명에서 확장자 및 언더스코어 제거
    cleaned_image_filename = row['이미지 파일명'].split('/')[-1].replace('.jpg', '').replace('_', ' ')

    if cleaned_image_filename == row["제목"]:
        img_src = row['이미지 파일명']
        html_str += f'<img src="{img_src}" width="100" alt="{row["제목"]}"/><br>'

# HTML 표시
HTML(html_str)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import pandas as pd
import os
import requests
import re

# ChromeOptions 설정
op = webdriver.ChromeOptions()
op.add_argument('--headless')
op.add_argument('--no-sandbox')
op.add_argument('--disable-dev-shm-usage')

def save_image_from_url(url, save_path):
    """이미지 URL에서 이미지를 다운로드하고 지정된 경로에 저장하는 함수"""
    response = requests.get(url)
    with open(save_path, 'wb') as f:
        f.write(response.content)

def clean_filename(filename):
    """파일명에서 허용되지 않는 문자를 제거하거나 교체하는 함수"""
    s = re.sub(r'[\\/*?:"<>|]', '', filename)  # 허용되지 않는 문자 제거
    return s.replace(' ', '_')  # 공백을 언더스코어대신 대체
    #return s.replace(' ', '')  # 공백을 언더스코어대신 공백으로 대체, None 이슈


# 크롤링 함수 정의
def crawl_kyobobook(item_id):
    url = f"https://product.kyobobook.co.kr/detail/{item_id}"

    # WebDriver 초기화 및 웹 페이지 접속
    driver = webdriver.Chrome(options=op)
    driver.get(url)

    # 웹 페이지 로딩을 기다립니다.
    wait = WebDriverWait(driver, 30)

    # 각 데이터의 CSS 선택자 혹은 XPath를 사용하여 데이터 추출
    title = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "span.prod_title"))).text
    author = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".author a:first-child"))).text

    try:
        editor_choice = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.info_text.fw_bold"))).text
    except TimeoutException:
        editor_choice = "N/A"

    book_intro = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.info_text"))).text
    contents = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "li.book_contents_item"))).text

    best_link_element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "a.btn_best_link")))
    best_link_text = best_link_element.text.replace("\n", ": ")

    price = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "span.prod_info_price > span.val"))).text + "원"
    rating = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "input.form_rating.rating-input"))).get_attribute("value") + "점"
    purchase_link = url
    genre1 = ' > '.join([elem.text for elem in driver.find_elements(By.CSS_SELECTOR, "li.category_list_item:first-child a")])
    genre2 = ' > '.join([elem.text for elem in driver.find_elements(By.CSS_SELECTOR, "li.category_list_item:last-child a")])


    # 이미지 URL 추출
    image_url = driver.find_element(By.CSS_SELECTOR, 'img[src^="https://contents.kyobobook.co.kr/sih/fit-in"]').get_attribute("src")

    # 이미지 다운로드 및 저장
    clean_title = clean_filename(title)
    image_filename = f"{clean_title}.jpg"
    image_save_path = f"/content/drive/MyDrive/ChatGpt 프로젝트/코딩/kogpt2/Crawling Data/이미지(책표지)/2/{image_filename}"
    save_image_from_url(image_url, image_save_path)
    cleaned_image_filename = image_filename.split('/')[-1].replace('.jpg', '').replace('_', ' ')

    # 브라우저 종료
    driver.quit()

    return {
        "제목": title,
        "저자": author,
        "편집장의 선택": editor_choice,
        "책소개": book_intro,
        "목차": contents,
        "주간베스트": best_link_text,
        "가격": price,
        "평점": rating,
        "구매링크": purchase_link,
        "장르1": genre1,
        "장르2": genre2,
        "이미지 파일명": cleaned_image_filename  #공백제거, image_save_path
    }

# 아이템 ID 목록
item_ids = [
    "S000202671445",
    "S000202266180",
    "S000000597147",
    "S000203331812",
    "S000201142283",
    "S000203244661",
    "S000061450857"
]

# 각 아이템 ID에 대해 크롤링 수행
results = [crawl_kyobobook(item_id) for item_id in item_ids]

# 데이터프레임 생성
df = pd.DataFrame(results)
display(df)
print()

# 저장할 파일명 지정
filename = "kyobobook_data5.csv"

# 지정된 경로에 파일 저장
path = f"/content/drive/MyDrive/ChatGpt 프로젝트/코딩/kogpt2/Crawling Data/{filename}"
df.to_csv(path, index=False)

print(f"'{filename}' 파일이 성공적으로 저장되었습니다.")
```
